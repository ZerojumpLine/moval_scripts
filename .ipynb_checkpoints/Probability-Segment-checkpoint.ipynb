```python
def caculate_dsc(score, inp, gt):
    """Calculate real dsc and estimated dsc

    Args:
        score: Confidence scores of a list of n ``(H, W, (D))``.
        inp: Network output of a list of n ``(d, H, W, (D))``
        gt: Ground truth segmentation of a list of n ``(H, W, (D))``.

    Return:
        estim_dsc: The mean estimated DSC for cases of shape ``(n,)``.
        real_dsc: The mean real DSC of for cases shape ``(n,)``.

    """
    from moval.models.utils import SoftDiceLoss
    num_class = inp[0].shape[0]
    estim_dsc_list = []
    #
    pred_all_flatten_bg = []
    gt_all_flatten_bg = []
    dsc = []
    for n_case in range(len(inp)):
        pred_case = np.argmax(inp[n_case], axis = 0) # ``(H, W, (D))``
        pred_flatten = pred_case.flatten() # ``n``
        score_case = score[n_case] # ``(H, W, (D))``
        score_flatten = score[n_case].flatten() # ``n``
        score_filled = np.zeros((score_flatten.shape + (num_class,))) # ``(n, d)``
        score_filled[np.arange(score_filled.shape[0]), pred_flatten] = score_flatten # ``(n, d)``
        score_filled = score_filled.T # ``(d, n)``
        score_filled = score_filled.reshape(((num_class,) + score_case.shape)) # ``(d, H, W, (D))``
        #
        estim_dsc = SoftDiceLoss(score_filled[np.newaxis, ...], pred_case[np.newaxis, ...])

        #
        gt_case     = gt[n_case] # ``(H, W, (D))``
        pos_bg = np.where(pred_case.flatten() == 0)[0]

        pred_all_flatten_bg.append(pred_case.flatten()[pos_bg])
        gt_all_flatten_bg.append(gt_case.flatten()[pos_bg])

        dsc_case = np.zeros(inp[n_case].shape[0])
        for kcls in range(1, inp[n_case].shape[0]):
            if np.sum(gt_case == kcls) == 0:
                estim_dsc[kcls] = -1
                dsc_case[kcls] = -1
            else:
                dsc_cal, _, _ = ComputMetric(pred_case == kcls, gt_case == kcls)
                dsc_case[kcls] = dsc_cal
        
        estim_dsc_list.append(np.mean(estim_dsc[1:][dsc_case[1:] >= 0]))
        dsc.append(np.mean(dsc_case[1:][dsc_case[1:] >= 0]))

    return estim_dsc_list, dsc
```

```python
def caculate_dsc_full(score, inp, gt):
    """Calculate real dsc and estimated dsc

    Args:
        score: Confidence scores of a list of n ``(H, W, (D))``.
        inp: Network output of a list of n ``(d, H, W, (D))``
        gt: Ground truth segmentation of a list of n ``(H, W, (D))``.

    Return:
        estim_dsc: The mean estimated DSC for cases of shape ``(n,)``.
        real_dsc: The mean real DSC of for cases shape ``(n,)``.

    """
    from moval.models.utils import SoftDiceLoss
    num_class = inp[0].shape[0]
    estim_dsc_list = []
    #
    pred_all_flatten_bg = []
    gt_all_flatten_bg = []
    dsc = []
    for n_case in range(len(inp)):
        pred_case = np.argmax(inp[n_case], axis = 0) # ``(H, W, (D))``
        pred_flatten = pred_case.flatten() # ``n``
        score_case = score[n_case] # ``(H, W, (D))``
        score_flatten = score[n_case].flatten() # ``n``
        score_filled = np.zeros((score_flatten.shape + (num_class,))) # ``(n, d)``
        score_filled[np.arange(score_filled.shape[0]), pred_flatten] = score_flatten # ``(n, d)``
        score_filled = score_filled.T # ``(d, n)``
        score_filled = score_filled.reshape(((num_class,) + score_case.shape)) # ``(d, H, W, (D))``
        #
        estim_dsc = SoftDiceLoss(score_filled[np.newaxis, ...], pred_case[np.newaxis, ...])

        #
        gt_case     = gt[n_case] # ``(H, W, (D))``
        pos_bg = np.where(pred_case.flatten() == 0)[0]

        pred_all_flatten_bg.append(pred_case.flatten()[pos_bg])
        gt_all_flatten_bg.append(gt_case.flatten()[pos_bg])

        dsc_case = np.zeros(inp[n_case].shape[0])
        for kcls in range(1, inp[n_case].shape[0]):
            if np.sum(gt_case == kcls) == 0:
                estim_dsc[kcls] = -1
                dsc_case[kcls] = -1
            else:
                dsc_cal, _, _ = ComputMetric(pred_case == kcls, gt_case == kcls)
                dsc_case[kcls] = dsc_cal
        
        estim_dsc_list.append(estim_dsc[1:])
        dsc.append(dsc_case[1:])

    return estim_dsc_list, dsc
```

```python
def lineplot(estim_dsc, dsc, title = "Estimation Error with MCP"):
    """Make the line plot to show the estimated segmentaiton error

    Args:
        estim_dsc: The estimated dsc for all the case with shape ``(n, )``
        dsc: The real dsc for all the case with shape ``(n, )``
        title: The given title to be shown.

    """

    d = {'Estimated DSC': estim_dsc, 'Real DSC': dsc}
    df = pd.DataFrame(data=d)
    df = df.sort_values(by="Real DSC", ascending=False, na_position='first')
    df = df.reset_index(drop=True)
    df.insert(2, "ID", range(len(dsc)), True)

    sns.set(rc={'figure.figsize':(12, 3)})
    sns.set_style("darkgrid")

    # Line plot
    ax = sns.lineplot(
        data=df,
        x="ID", y="Estimated DSC",
        color='black', lw=3, alpha=0.7,
        label='Estimated DSC'
    )

    sns.scatterplot(
        data=df,
        x="ID", y="Estimated DSC",
        color='black',
        marker='o',
        s=50,
        label=None
    )

    # Fill between the lines
    plt.fill_between(
        df['ID'], 0, df['Real DSC'],
        color='grey', alpha=0.5, label='Real DSC'
    )
    
    # Fill between the lines
    plt.fill_between(
        df['ID'],
        df['Estimated DSC'],
        df['Real DSC'],
        color='#A2142F',
        alpha=0.5,
        label='Estimatation Gap'
    )

    plt.ylim(0., 1)
    plt.xlim(-1, len(df))
    
    ax.set_xticks([])
    ax.set_xlabel('Different Testing Samples')

    # Add legend
    plt.legend(bbox_to_anchor=(0.5, -0.1), loc='upper left', ncols=3)

    plt.title(title)

    plt.savefig(f'Fig_{title}.pdf', bbox_inches='tight')
    plt.show()
```

```python
def lineplot_whole(estim_dsc, dsc, title):
    # show the overall perf
    # _mean_conf
    # acc_all
    sns.set(rc={'figure.figsize':(1,3)})
    sns.set_style("darkgrid")
    
    _mean_conf = np.mean(estim_dsc)
    acc_all = np.mean(dsc)
    
    x_left = 0
    x_right = 1
    
    ax = plt.gca()
    
    y_down = np.min((_mean_conf, acc_all))
    y_up = np.max((_mean_conf, acc_all))
    ax.fill_between([x_left, x_right], y_down, y_up, linewidth=0, edgecolor='none', facecolor='#A2142F', alpha=0.5)    
    # Create a hatched texture fill for the box with gradient color
    gradient_steps = 50  # Number of steps for the gradient
    for i in range(gradient_steps):
        # Calculate the gradient color at this step
        gradient_color = (1 - i / gradient_steps) * np.array([1, 1, 1]) + (i / gradient_steps) * np.array([0x00, 0x00, 0x00]) / 255
        facecolor = tuple(gradient_color)
        
        # Add a rectangle patch for this step
        x_start = (i / gradient_steps) * acc_all
        x_end = ((i + 1) / gradient_steps) * acc_all
        texture_box = patches.Rectangle((x_left, x_start), x_right - x_left, x_end - x_start, linewidth=0, edgecolor='none', facecolor=facecolor, alpha=0.5)
        ax.add_patch(texture_box)
    
    
    plt.axhline(_mean_conf, color="black", lw=3, alpha=0.7, ls='-')
    
    plt.ylim(0., 1)
    
    # Remove x-axis ticks
    plt.xticks([])
    
    plt.title(title, y=1.05)  # Adjust fontsize and y position as needed
    plt.savefig(f'Fig_{title}.pdf', bbox_inches='tight')
    
    plt.show()
```

```python
predpath = cardiac_predpath
gtpath = cardiac_gtpath
predfiles = [f for f in listdir(predpath) if isfile(join(predpath, f))]

logits = []
gt = []
imgs_test = []
for predfile in predfiles[-36:]:
    if predfile.split('_')[-2][-1] == '0':
        print(predfile)
        # grep the caseID
        caseID = f"{predfile.split('_')[-3]}_{predfile.split('_')[-2][:-4]}"
        #
        GT_file = f"{gtpath}/{caseID}/seg.nii.gz"
        #
        logit_cls0_file = f"{predpath}/pred_{caseID}cls0_prob.nii.gz"
        logit_cls1_file = f"{predpath}/pred_{caseID}cls1_prob.nii.gz"
        logit_cls2_file = f"{predpath}/pred_{caseID}cls2_prob.nii.gz"
        logit_cls3_file = f"{predpath}/pred_{caseID}cls3_prob.nii.gz"
        #
        logit_cls0_read = nib.load(logit_cls0_file)
        logit_cls1_read = nib.load(logit_cls1_file)
        logit_cls2_read = nib.load(logit_cls2_file)
        logit_cls3_read = nib.load(logit_cls3_file)
        #
        logit_cls0      = logit_cls0_read.get_fdata()   # ``(H, W, D)``
        logit_cls1      = logit_cls1_read.get_fdata()
        logit_cls2      = logit_cls2_read.get_fdata()
        logit_cls3      = logit_cls3_read.get_fdata()
        #
        img_file = f"/well/win-fmrib-analysis/users/gqu790/moval/Robust-Medical-Segmentation/data/Dataset_Cardiac/3/{caseID}/image.nii.gz"
        img_read = nib.load(img_file)
        img_data      = img_read.get_fdata()   # ``(H, W, D)``
        #
        GT_read         = nib.load(GT_file)
        GTimg           = GT_read.get_fdata()           # ``(H, W, D)``
        #
        logit_cls = np.stack((logit_cls0, logit_cls1, logit_cls2, logit_cls3))  # ``(d, H, W, D)``
        # only including the slices that contains labels
        for dslice in range(GTimg.shape[2]):
            if np.sum(GTimg[:, :, dslice]) > 0:
                logits.append(logit_cls[:, :, :, dslice])
                gt.append(GTimg[:, :, dslice])
                imgs_test.append(img_data[:, :, dslice])
```

```python
def visual_2Dmap(conf_maps_0, conf_maps_1, conf_maps_2, imgs_toshow, lbls_toshow, preds_toshow, _real_dscs, _estim_dscs_0, _estim_dscs_1, _estim_dscs_2, title = 'The Visualization of MCP for Cases with High and Low DSC'):
    """Visualization of confidence score for two cases.

    Args:
        conf_maps: A list of confidence maps, of shape ``(H, W)``.
        imgs_toshow: A list of image, of shape ``(H, W)``.
        lbls_toshow: A list of label, of shape ``(H, W)``.
        preds_toshow: A list of prediction, of shape ``(H, W)``.
        _real_dscs: A list of real DSC (float).
        _estim_dscs: A list of esimtaed DSC (float).
        title: The given title to be shown.

    """

    from matplotlib.gridspec import GridSpec

    fig = plt.figure(figsize=(15, 8))
    gs = GridSpec(2, 3, width_ratios=[1, 1, 1])  # Adjust the ratios as needed

    cmap = sns.color_palette("ch:s=-.2,r=.6", as_cmap=True)

    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])
    ax2 = plt.subplot(gs[2])
    ax3 = plt.subplot(gs[3])
    ax4 = plt.subplot(gs[4])
    ax5 = plt.subplot(gs[5])

    xleft = round(imgs_toshow[0].shape[1]/2) - 80
    xright = round(imgs_toshow[0].shape[1]/2) + 80
    yleft = round(imgs_toshow[0].shape[0]/2) - 80
    yright = round(imgs_toshow[0].shape[0]/2) + 80

    # Define colors for each class
    class_colors = ['', 'green', 'blue', 'red']  # Add more colors if you have more classes

    ax0.imshow(imgs_toshow[0].T[xleft:xright, yleft:yright], cmap='gray')
    # Overlay the label map using class-specific colors
    for i in range(len(class_colors)):
        if i == 0:  # Skip label 0
            continue
        mask = lbls_toshow[0] == i
        label_map_rgb = np.zeros((*lbls_toshow[0].shape, 4))
        label_map_rgb[mask, :] = plt.cm.colors.to_rgba(class_colors[i])
        ax0.imshow(np.transpose(label_map_rgb, (1, 0, 2))[xleft:xright, yleft:yright], alpha=0.5)
    ax0.set_xticks([])
    ax0.set_yticks([])
    ######
    ax1.imshow(imgs_toshow[0].T[xleft:xright, yleft:yright], cmap='gray')
    # Overlay the label map using class-specific colors
    for i in range(len(class_colors)):
        if i == 0:  # Skip label 0
            continue
        mask = preds_toshow[0] == i
        label_map_rgb = np.zeros((*preds_toshow[0].shape, 4))
        label_map_rgb[mask, :] = plt.cm.colors.to_rgba(class_colors[i])
        ax1.imshow(np.transpose(label_map_rgb, (1, 0, 2))[xleft:xright, yleft:yright], alpha=0.5)
    ax1.set_xticks([])
    ax1.set_yticks([])
    ######
    diff = np.abs(preds_toshow[0] - lbls_toshow[0])
    sns.heatmap(diff.T[xleft:xright, yleft:yright], cmap=cmap, xticklabels=False, yticklabels=False, cbar_kws={'format': ''}, ax=ax2)
    cbar = ax2.collections[0].colorbar
    cbar.set_ticks([])
    ######
    unc = 1 - conf_maps_0[0]
    sns.heatmap(unc.T[xleft:xright, yleft:yright], cmap=cmap, xticklabels=False, yticklabels=False, cbar_kws={'format': ''}, ax=ax3)
    cbar = ax3.collections[0].colorbar
    cbar.set_ticks([])
    ######
    unc = 1 - conf_maps_1[0]
    sns.heatmap(unc.T[xleft:xright, yleft:yright], cmap=cmap, xticklabels=False, yticklabels=False, cbar_kws={'format': ''}, ax=ax4)
    cbar = ax4.collections[0].colorbar
    cbar.set_ticks([])
    ###### 
    unc = 1 - conf_maps_2[0]
    sns.heatmap(unc.T[xleft:xright, yleft:yright], cmap=cmap, xticklabels=False, yticklabels=False, cbar_kws={'format': ''}, ax=ax5)
    cbar = ax5.collections[0].colorbar
    cbar.set_ticks([])
    #
    ax0.set_title('Ground Truth Segmentation\n[Green: 1, Blue: 2, Red: 3]', x=0.5, y=1)
    ax1.set_title('Predicted Segmentation', x=0.5, y=1.05)
    ax2.set_title('Difference between Prediction and Ground Truth', x=0.5, y=1.1)
    ax3.set_title('MCP', x=0.5, y=1.1)
    ax4.set_title('TS MCP', x=0.5, y=1.1)
    ax5.set_title('MOVAL Ensemble', x=0.5, y=1.1)
    #
    ax2.text(0.5, 1.05,  f'Real DSC is [{_real_dscs[0][0]:.3f}, {_real_dscs[0][1]:.3f}, {_real_dscs[0][2]:.3f}]', transform=ax2.transAxes, ha='center', va='center', fontsize=12)
    ax3.text(0.5, 1.05,  f'Estimated DSC is [{_estim_dscs_0[0][0]:.3f}, {_estim_dscs_0[0][1]:.3f}, {_estim_dscs_0[0][2]:.3f}]', transform=ax3.transAxes, ha='center', va='center', fontsize=12)
    ax4.text(0.5, 1.05,  f'Estimated DSC is [{_estim_dscs_1[0][0]:.3f}, {_estim_dscs_1[0][1]:.3f}, {_estim_dscs_1[0][2]:.3f}]', transform=ax4.transAxes, ha='center', va='center', fontsize=12)
    ax5.text(0.5, 1.05,  f'Estimated DSC is [{_estim_dscs_2[0][0]:.3f}, {_estim_dscs_2[0][1]:.3f}, {_estim_dscs_2[0][2]:.3f}]', transform=ax5.transAxes, ha='center', va='center', fontsize=12)

    fig.suptitle(title, fontsize=16, y=0.98)

    plt.savefig(f'Fig_{title}.pdf', bbox_inches='tight')
    plt.show()
```

```python
def make_3d_plot(conf_map_0, conf_map_1, conf_map_2, img_toshow, gt_toshow, pred_toshow, _real_dsc, _estim_dsc_0, _estim_dsc_1, _estim_dsc_2, title = 'The Visualization of MCP'):
    """Visualization of confidence score using 3D scatter plot.

    Args:
        conf_map_0: The calculated 3D confidence maps, of shape ``(H, W, D)``.
        img_toshow: An example 2D image, of shape ``(H, W)``.
        gt_toshow: The corresponding 3D label map, of shape ``(H, W, D)``.
        pred_toshow:The corresponding 3D prediction, of shape ``(H, W, D)``.
        _real_dsc (float): The real DSC calculated by the 3D volume.
        _estim_dsc (float): The esimtaed DSC.
        title: The given title to be shown.

    """

    # Set up the 3D plot
    import matplotlib.patches as mpatches

    sns.set_style("white")
    fig, axes = plt.subplots(2, 3, figsize=(15, 8), subplot_kw={'projection': '3d', 'computed_zorder': False})

    fig.subplots_adjust(wspace=0.)  # Adjust the value as needed

    ############################
    threshold = 0.1  # Adjust as needed
    # Plot each 2D scatter plot as a separate layer
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = conf_map_0[:, :, i]
        # Filter out small values
        mask = z > threshold
        x = x[mask]
        y = y[mask]
        z = z[mask]
        axes[1, 0].scatter(x, y, i * np.ones_like(x), c='#99687f', alpha=0.005, zorder=1)
    # Get the maximum z value from the scatter plot
    max_z_scatter = 0
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[1, 0].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    # Remove tick lines and labels
    axes[1, 0].set_title('MCP')

    ############################
    threshold = 0.1  # Adjust as needed
    # Plot each 2D scatter plot as a separate layer
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = conf_map_1[:, :, i]
        # Filter out small values
        mask = z > threshold
        x = x[mask]
        y = y[mask]
        z = z[mask]
        axes[1, 1].scatter(x, y, i * np.ones_like(x), c='#99687f', alpha=0.005, zorder=1)
    # Get the maximum z value from the scatter plot
    max_z_scatter = 0
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[1, 1].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    # Remove tick lines and labels
    axes[1, 1].set_title('TS MCP')

    # ############################
    threshold = 0.1  # Adjust as needed
    # Plot each 2D scatter plot as a separate layer
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = conf_map_2[:, :, i]
        # Filter out small values
        mask = z > threshold
        x = x[mask]
        y = y[mask]
        z = z[mask]
        axes[1, 2].scatter(x, y, i * np.ones_like(x), c='#99687f', alpha=0.005, zorder=1)
    # Get the maximum z value from the scatter plot
    max_z_scatter = 0
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[1, 2].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    # Remove tick lines and labels
    axes[1, 2].set_title('MOVAL Ensemble')

    # ############################
    # Plot each 2D scatter plot as a separate layer
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = gt_toshow[:, :, i]
        # Filter out small values
        mask = z == 1
        x = x[mask]
        y = y[mask]
        z = z[mask]

        axes[0, 0].scatter(x, y, i * np.ones_like(x), c='green', alpha=0.05, zorder=1)
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[0, 0].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    axes[0, 0].set_title('Ground Truth Segmentation')

    # ############################
    # Define a threshold for filtering out small values
    # Plot each 2D scatter plot as a separate layer
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = np.abs(pred_toshow[:, :, i] - gt_toshow[:, :, i])
        # Filter out small values
        mask = z == 1
        x = x[mask]
        y = y[mask]
        z = z[mask]
        axes[0, 2].scatter(x, y, i * np.ones_like(x), c='#99687f', alpha=0.5, zorder=1)
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[0, 2].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    axes[0, 2].set_title('Difference between Prediction and Ground Truth')

    # ############################
    # Define a threshold for filtering out small values
    for i in range(conf_map_0.shape[2]):
        x, y = np.meshgrid(np.arange(conf_map_0.shape[1]), np.arange(conf_map_0.shape[0]))
        z = pred_toshow[:, :, i]
        # Filter out small values
        mask = z == 1
        x = x[mask]
        y = y[mask]
        z = z[mask]

        axes[0, 1].scatter(x, y, i * np.ones_like(x), c='green', alpha=0.05, zorder=1)
    # Show the image on the bottom x-y plane
    x, y = np.meshgrid(np.arange(img_toshow.shape[1]), np.arange(img_toshow.shape[0]))
    axes[0, 1].plot_surface(x, y, max_z_scatter * np.ones_like(img_toshow), facecolors=plt.cm.gray(img_toshow), rstride=5, cstride=5, shade=False, zorder=0)
    axes[0, 1].set_title('Predicted Segmentation')

    print('Finish plotting...')

    for y_axis in range(2):
        for k_axis in range(3):
            axes[y_axis, k_axis].set_xticks([])
            axes[y_axis, k_axis].set_yticks([])
            axes[y_axis, k_axis].set_zticks([])
            axes[y_axis, k_axis].set_xticklabels([])
            axes[y_axis, k_axis].set_yticklabels([])
            axes[y_axis, k_axis].set_zticklabels([])
            axes[y_axis, k_axis].set_xlim(0, 99)
            axes[y_axis, k_axis].set_ylim(0, 99)
            axes[y_axis, k_axis].xaxis.set_pane_color((0.850, 0.816, 0.686))
            axes[y_axis, k_axis].yaxis.set_pane_color((0.850, 0.816, 0.686))
            axes[y_axis, k_axis].zaxis.set_pane_color((0.850, 0.816, 0.686))
    
            axes[y_axis, k_axis].grid(False)

    alpha_patch = mpatches.Patch(color='#99687f', alpha = 0.5, label='Uncertain Region')
    red_patch = mpatches.Patch(color='#99687f', label='Segmentation Error')
    green_patch = mpatches.Patch(color='green', label='Segmentation Label')
    axes[0, 0].legend(handles=[green_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')
    axes[0, 1].legend(handles=[green_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')
    axes[0, 2].legend(handles=[red_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')
    axes[1, 0].legend(handles=[alpha_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')
    axes[1, 1].legend(handles=[alpha_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')
    axes[1, 2].legend(handles=[alpha_patch], bbox_to_anchor=(0.15, 0.05), loc='upper left')

    axes[0, 2].text2D(0.5, 0.96,  f'Real DSC is {_real_dsc:.3f}', transform=axes[0, 2].transAxes, ha='center', va='center', fontsize=12)
    axes[1, 0].text2D(0.5, 0.96,  f'Estimated DSC is {_estim_dsc_0:.3f}', transform=axes[1, 0].transAxes, ha='center', va='center', fontsize=12)
    axes[1, 1].text2D(0.5, 0.96,  f'Estimated DSC is {_estim_dsc_1:.3f}', transform=axes[1, 1].transAxes, ha='center', va='center', fontsize=12)
    axes[1, 2].text2D(0.5, 0.96,  f'Estimated DSC is {_estim_dsc_2:.3f}', transform=axes[1, 2].transAxes, ha='center', va='center', fontsize=12)

    fig.suptitle(title, fontsize=16, y=0.95)
    # Show the plot

    plt.savefig(f'Fig_{title}.pdf', bbox_inches='tight')
    print('Figure saved.')
    plt.show()
```
